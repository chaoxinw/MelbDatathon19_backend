```python
temp_17_df = pd.read_csv("temp_17_proserpine.csv")
for c in temp_17_df.columns:
    temp_17_df.fillna((temp_17_df[c].mean()), inplace=True)


temp_18_df = pd.read_csv("temp_18_proserpine.csv")
for c in temp_18_df.columns:
    temp_18_df.fillna((temp_18_df[c].mean()), inplace=True)


temp_19_df = pd.read_csv("temp_19_proserpine.csv")
for c in temp_19_df.columns:
    temp_19_df.fillna((temp_19_df[c].mean()), inplace=True)
```

```python
with open("../servercodes/T55KFT_mean_ndvi_total_2.json", "r") as fp:
    ndvi_dict = json.load(fp)

date_list = ndvi_dict["date_list"]
```

```python
d_list = date_list[27:]
```

```python
temp_list = []
for d in tqdm(d_list):
    do = datetime.strptime(d, "%Y-%m-%d")
    mon = do.strftime("%b")
    day = do.day
    year = do.year
    if year == 2016:
        temp = temp_16_df[mon].values.tolist()[day-1]
    elif year == 2017:
        temp = temp_17_df[mon].values.tolist()[day-1]
    elif year == 2018:
        temp = temp_18_df[mon].values.tolist()[day-1]
    elif year == 2019:
        temp = temp_19_df[mon].values.tolist()[day-1]
#     print(type(do.year))
    temp_list.append(temp)
```

```python
# same with rainfall
rainfall_16_df = pd.read_csv("rainfall_16_proserpine.csv")
for c in rainfall_16_df.columns:
    rainfall_16_df.fillna((rainfall_16_df[c].mean()), inplace=True)

rainfall_17_df = pd.read_csv("rainfall_17_proserpine.csv")
for c in rainfall_17_df.columns:
    rainfall_17_df.fillna((rainfall_17_df[c].mean()), inplace=True)


rainfall_18_df = pd.read_csv("rainfall_18_proserpine.csv")
for c in rainfall_18_df.columns:
    rainfall_18_df.fillna((rainfall_18_df[c].mean()), inplace=True)


rainfall_19_df = pd.read_csv("rainfall_19_proserpine.csv")
for c in rainfall_19_df.columns:
    rainfall_19_df.fillna((rainfall_19_df[c].mean()), inplace=True)
```

```python
rainfall_list = []
for d in tqdm(d_list):
    do = datetime.strptime(d, "%Y-%m-%d")
    mon = do.strftime("%b")
    day = do.day
    year = do.year
    if year == 2016:
        rainfall = rainfall_16_df[mon].values.tolist()[day-1]
    elif year == 2017:
        rainfall = rainfall_17_df[mon].values.tolist()[day-1]
    elif year == 2018:
        rainfall = rainfall_18_df[mon].values.tolist()[day-1]
    elif year == 2019:
        rainfall = rainfall_19_df[mon].values.tolist()[day-1]
#     print(type(do.year))
    rainfall_list.append(rainfall)
```

```python
total_train_df = pd.read_csv("../servercodes/timeseries_traindf.csv")

total_train_df["temp"] = temp_list
total_train_df["rainfall"] = rainfall_list

total_train_df.head()
```

```python
is_drought_list = []
mean_ndvi_list = total_train_df["mean_ndvi"].values.tolist()
for i in list(range(len(temp_list))):
    temp = temp_list[i]
    rainfall = rainfall_list[i]
    
    if temp > 37 and rainfall <= 0.5:
        is_drought_list.append(1)
        print("drought", mean_ndvi_list[i])
    else:
        is_drought_list.append(0)
        
    
```

```python
total_train_df["drought"] = is_drought_list
```
